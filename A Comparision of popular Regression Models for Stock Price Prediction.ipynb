# Splitting our features for testing and training

```python
x_train,x_test,y_train,y_test = df_train.drop('Close',axis=1),df_test.drop('Close',axis=1),df_train['Close'],df_test['Close']
```

# Implimenting our models

## Multiple Linear Regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

## Support Vector Regression

### SVM is not good at handling unscaled data, so we will try to scale our features

```python
x_svm_train,x_svm_test,y_svm_train,y_svm_test = df_train.drop(['Close'],axis=1),df_test.drop(['Close'],axis=1),df_train['Close'],df_test['Close']
```

### Scaling our features

```python
scaler = prep.MaxAbsScaler()
scaler.fit(x_svm_train)
```

```python
x_svm_train[x_svm_train.columns] = scaler.transform(x_svm_train)
x_svm_test[x_svm_test.columns] = scaler.transform(x_svm_test)
```

### Implimenting our model

```python
svr = SVR(kernel='rbf',C=100,epsilon=.01)
svr.fit(x_svm_train,y_svm_train)
```

## Decision Tree Regression

```python
dtr = DecisionTreeRegressor(criterion='absolute_error',max_depth=6)
dtr.fit(x_train,y_train)
```

## Random Forest Regression

```python
rfr = RandomForestRegressor(criterion='absolute_error',max_depth=6,n_estimators=8,n_jobs=-1)
rfr.fit(x_train,y_train)
```

# Comparing our accuracy scores

## Getting our accuracy scores

```python
data_train = [
    round(reg.score(x_train,y_train)*100,2),
    round(svr.score(x_svm_train,y_train)*100,2),
    round(dtr.score(x_train,y_train)*100,2),
    round(rfr.score(x_train,y_train)*100,2),
]
data_test = [
    round(reg.score(x_test,y_test)*100,2),
    round(svr.score(x_svm_test,y_test)*100,2),
    round(dtr.score(x_test,y_test)*100,2),
    round(rfr.score(x_test,y_test)*100,2),
]
```

## Making a dataframe out of it

```python
idx = ['Multiple Linear Regression','Support Vector Regression','Decision Tree Regression','Random Forest Regression']
scores = pd.DataFrame({
    "Training Score" : data_train,
    "Testing Score" : data_test
},index=idx)
scores
```

### Sorting it according to test scores

```python
scores.sort_values(by='Testing Score',ascending=False,inplace=True)
scores
```

# With this we can conclude that
## Decision Tree Regression is the best as it handles unseen data the best and is less resource intensive
## Random Forest Regression follows it closely but is more resuorce intensive
## Multiple Linear Regression is the best when it comes to training set but underperformed a little on the testing set
## Support Vector Machine underperforms slightly as it is bad for huge data
